```python
# import libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#load dataset from url
df=pd.read_csv('https://raw.githubusercontent.com/fslaborg/datasets/main/data/iris.csv')
```

```python
df.head(5)
```

```python
#check shape(rowsx column)
df.shape
```

```python
#check info about dataset
df.info()
```

```python
#Basics statistics
df.describe()
```

```python
#check missing values
df.isnull().sum()
```

```python
# count how many samples per species
df['species'].value_counts()
```

```python
# group by species and find mean
df.groupby('species').mean()
```

```python
#Visulaizationsn using histogram
df.hist(figsize=(10,6))
plt.suptitle('Feature Distribution ')
plt.show()
```

```python
#visualization using pairplot
sns.pairplot(df,hue='species')
plt.show()
```

```python
#correlation Heatmap
plt.figure(figsize=(6,4))
# Drop the 'species' column before calculating the correlation
sns.heatmap(df.drop('species', axis=1).corr(), annot=True, cmap='coolwarm')
plt.show()
```
